```python
#Q4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

#Take the input from the user
x=input("Enter the item you wish to search in flipkart.com  ")

#opening up www.flipkart.com website on automated chrome window
driver.get('https://www.flipkart.com/')

#finding element in search bar
search_field=driver.find_element_by_class_name("_3704LK") #search bar
search_field.send_keys(x)

#clicking on search icon
search_icon=driver.find_element_by_xpath("/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/button") #location search
search_icon.click()

brand=[]                   #list to store brand name
smartphone=[]              #list to store smartphone name
colour=[]                  #list to store colour
RAM=[]                     #list to store RAM
storage=[]                 #list to store storage
primary_camera=[]          #list to store primary camera
secondary_camera=[]        #list to store secondary camera
display_size=[]            #list to store display size
battery_capacity=[]        #list to store battery capacity
price=[]                   #list to store price
product_url=[]             #list to store product url

time.sleep(5)
#Lets extract all the tags having the brand name
try:
    brand_tags=driver.find_elements_by_xpath("//div[@class='_4rR01T']") #locating web element of brand
    for i in brand_tags:
        titles=i.text.split()[0]                                           #Now the text of the brand name is inside the tag extracted above
        brand.append(''.join(titles))
    
except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
    brand.append("-")   
    
#Lets extract all the tags having the smartphone name
try:                                       
    sp_tags=driver.find_elements_by_xpath("//div[@class='_4rR01T']") #locating web element of smartphone
    for i in sp_tags:
        lm=len(i.text.split())
        for j in range(0,lm):
            a=i.text.split()[j]
            if a[0]=='(':
                z=j
                break
        sp=i.text.split()[1:j]                                          #Now the text of the smartphone name is inside the tag extracted above
        smartphone.append(' '.join(sp))
    
except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
        smartphone.append("-") 

#Lets extract all the tags having the colour
try:
    color_tags=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
    for i in color_tags:
        kl=len(i.text.split())
        for jk in range(0,kl):
            a=i.text.split()[jk]
            if a[0]=='(':
                zi=jk
                break
        col=i.text.replace('(','').split()[zi]                                          #Now the text of the colour is inside the tag extracted above
        colour.append(''.join(col))
    
except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
    colour.append("-")
    
#Lets extract all the tags having the RAM
try:
    ram_tags=driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[1]") #locating web element of RAM
    for i in ram_tags:
        r=i.text.split()[0:2]                                           #Now the text of the RAM is inside the tag extracted above
        RAM.append(' '.join(r))
    
except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
    RAM.append("-") 

#Lets extract all the tags having the storage
try:
    store_tags=driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[1]") #locating web element of storage
    for i in store_tags:
        s=i.text.split()[4:6]                                           #Now the text of the storage is inside the tag extracted above
        storage.append(' '.join(s))
    
except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
    storage.append("-") 
    
#Lets extract all the tags having the primary camera
try:
    prim_tags=driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[3]") #locating web element of primary camera
    for i in prim_tags:
        prim=i.text.split()[0]                                           #Now the text of the primary camera is inside the tag extracted above
        primary_camera.append(''.join(prim))
    
except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
    primary_camera.append("-") 
    
#Lets extract all the tags having the display size
try:
    disp_tags=driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[2]") #locating web element of display size
    for i in disp_tags:
        disp=i.text.split()[0:2]                                           #Now the text of the display size is inside the tag extracted above
        display_size.append(' '.join(disp))
    
except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
    display_size.append("-") 

#Lets extract all the tags having the battery capacity
try:
    batcap_tags=driver.find_elements_by_xpath("//ul[@class='_1xgFaf']/li[4]") #locating web element of battery capacity
    for i in batcap_tags:
        bat=i.text.split()[0:2]                                           #Now the text of the battery capacity is inside the tag extracted above
        battery_capacity.append(' '.join(bat))
    
except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
    battery_capacity.append("-")

#Lets extract all the tags having the price
try:
    price_tags=driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']") #locating web element of price
    for i in price_tags:
        prices=i.text                                           #Now the text of the price is inside the tag extracted above
        price.append(prices)
    
except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
    price.append("-")
    
#Lets extract all the tags having the product url
try:
    prod_tags=driver.find_elements_by_xpath("//a[@class='_1fQZEK']") #locating web element of product url
    for i in prod_tags:
        pr=i.get_attribute('href')                                          #Now the text of the product url is inside the tag extracted above
        product_url.append(pr)
    
except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
    product_url.append("-")
    
l=len(product_url)
for b in range(0,l):
    
    driver.get(product_url[b])                         #opening each product url to retrieve further information
    
    time.sleep(5) 
    
    #Lets extract all the tags having the secondary camera
    try:
        secon_tags=driver.find_elements_by_xpath("//div[@class='_2418kt']/ul/li[3]") #locating web element of secondary camera
        for i in secon_tags:
            secon=i.text.split()[0]                                           #Now the text of the secondary camera is inside the tag extracted above
            secondary_camera.append(''.join(secon))
    
    except NoSuchElementException:                              #if element not traceable/ data unavailble replace with '-'
        secondary_camera.append("-") 

pd.set_option("display.max_rows", None, "display.max_columns", None)                #To display the entire dataframe
df4=pd.DataFrame({'Brand Name':brand, 'Smartphone Name':smartphone, 'Colour': colour, 'RAM':RAM, 'Storage':storage, 'Primary Camera':primary_camera, 'Secondary Camera':secondary_camera, 'Storage':storage, 'Display Size':display_size, 'Battery Capacity':battery_capacity, 'Price':price, 'Product URL':product_url})
df4.to_csv('Flipkart Data')
print("The data for ", x)
df4
```

```python
#Q10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

#opening up https://www.hostelworld.com/ in London location seting checkin date as 27 July and check out date as 30 July
driver.get('https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-07-27&to=2022-07-30&guests=2&HostelNumber=&page=1')

time.sleep(5)

hostel_name=[]     #list to store hostel name
distance=[]        #list to store distance from city centre 
ratings=[]         #list to store ratings 
total_reviews=[]   #list to store total reviews 
overall_reviews=[] #list to store overall reviews, 
price_p=[]         #list to store privates from price
price_d=[]         #list to store dorms from price 
facilities=[]      #list to store facilities 
prop=[]            #list to store property description.
url=[]

for j in range(0,3):
    time.sleep(5)
    #Lets extract all the tags having the hostel_name
    hostel_tags=driver.find_elements_by_xpath("//h2[@class='title title-6']") #locating web element of hostel name
    for i in hostel_tags:
        if i.text is None:
            hostel_name.append("--")
        else:
            hostel_name.append(i.text)
    
    #Lets extract all the tags having the distance from city centre
    dis=driver.find_elements_by_xpath("//span[@class='description']") #locating web element of distance from city centre
    for i in dis:
        if i.text is None:
            distance.append("--")
        else:
            distance.append(i.text)

    #Lets extract all the tags having the rating
    rat_tag=driver.find_elements_by_xpath("//div[@class='rating rating-summary-container big']/div[1]") #locating web element of rating
    for i in rat_tag:
        if i.text is None:
            ratings.append("--")
        else:
            ratings.append(i.text)

    #Lets extract all the tags having the total reviews
    tot_rev=driver.find_elements_by_xpath("//div[@class='reviews']") #locating web element of total reviews
    for i in tot_rev:
        if i.text is None:
            total_reviews.append("--")
        else:
            total_reviews.append(i.text)
    
    #Lets extract all the tags having the overall reviews
    ovr_rev=driver.find_elements_by_xpath("//div[@class='rating rating-summary-container big']/div[2]/div[1]/span") #locating web element of overall reviews
    for i in ovr_rev:
        if i.text is None:
            overall_reviews.append("--")
        else:
            overall_reviews.append(i.text)
    
    #Lets extract all the tags having the privates from price
    pp_tag=driver.find_elements_by_xpath("//div[@class='prices-col']/a/div[1]") #locating web element of privates from price
    for i in pp_tag:
        if i.text is None:
            price_p.append("--")
        else:
            price_p.append(i.text)

    
    #Lets extract all the tags having the dorms from price
    dp_tag=driver.find_elements_by_xpath("//div[@class='prices-col']/a/div[2]") #locating web element of dorms from price
    for i in dp_tag:
        if i.text is None:
            price_d.append("--")
        else:
            price_d.append(i.text)

    
    #Lets extract all the tags having the urls
    u_tag=driver.find_elements_by_xpath("//a[@class='view-button']") #locating web element of url
    for i in u_tag:
        uv=i.get_attribute('href')                                          #Now the text of the urls is inside the tag extracted above
        url.append(uv)
    
    try:
        #Now clicking on the next button
        time.sleep(5)                                                                                                                     
        next_button=driver.find_element_by_xpath("/html/body/div[3]/div/div/div[1]/div[4]/div/div/div[33]/div[6]")
        next_button.click()
        time.sleep(5)
    except NoSuchElementException:
        break
        
l=len(url)    
for b in range(0,l):
    driver.get(url[b])                         #opening each product url to retrieve further information
    time.sleep(10)
    
    #Lets extract all the tags having the property description
    pt_tag=driver.find_elements_by_xpath("//div[@class='description-container']/div/div[2]") #locating web element of property description
    z=1
    for i in pt_tag:
        pt=i.text         #Now the text of the property description is inside the tag extracted above
        if z==1:
            prop.append(pt)
            z=z+1
    
    time.sleep(5)
    fac=[]
    #Lets extract all the tags having the facilities
    fc_tag=driver.find_elements_by_xpath("//ul[@class='facility-sections']/li") #locating web element of facilities
    for i in fc_tag:
        fc=i.text           #Now the text of the facilities is inside the tag extracted above
        fac.append(fc)
    facilities.append(' '.join(fac))
        

pd.set_option("display.max_rows", None, "display.max_columns", None)                #To display the entire dataframe
df10=pd.DataFrame({'Hostel Name ':hostel_name, 'Distance from city centre':distance, 'Ratings': ratings, 'Total reviews':total_reviews, 'Overall reviews':overall_reviews, 'Privates from price':price_p, 'Dorms from price':price_d, 'Facilities':facilities, 'Property description':prop })
print("The data for all available Hostels from https://www.hostelworld.com/ in “London” location.")
df10
```
