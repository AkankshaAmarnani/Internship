```python
#Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

#opening up naukri.com website on automated chrome window
driver.get('https://www.naukri.com/')

#finding element for job search bar
search_field_designation=driver.find_element_by_class_name("suggestor-input ") #job search bar
search_field_designation.send_keys("Data Analyst")

#entering location Bangalore in location search bar 
search_field_location=driver.find_element_by_xpath("/html/body/div/div[2]/div[3]/div/div/div[5]/div/div/div/input") #location search
search_field_location.send_keys("Bangalore")

#clicking on search button
search_button=driver.find_element_by_xpath("/html/body/div/div[2]/div[3]/div/div/div[6]") #location search
search_button.click()

job_titles=[]                     #list to store job titles
company_names=[]                  #list to store company names
loc=[]                            #list to store location
experience_list=[]                #list to store experience
    
#Lets extract all the tags having the job titles
title_tags=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']") #locating web element of title

#Now the text of the job title is inside the tag extracted above
for i in title_tags[0:10]:
    titles=i.text
    job_titles.append(titles)

#Lets extract all the tags having the location
locations_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]") #locating web element of company name

#Now the location is inside the tag extracted above
for i in locations_tags[0:10]:
    location=i.text
    loc.append(location)

#Lets extract all the tags having the company names
companies_tags=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']") #locating web element of company name

#Now the text of the company name is inside the tag extracted above
for i in companies_tags[0:10]:
    company_name=i.text
    company_names.append(company_name)

#Lets extract all the tags having the experience requirement
experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span") #locating web element of company name

#Now the experience requirement is inside the tag extracted above
for i in experience_tags[0:10]:
    experience=i.text
    experience_list.append(experience)

pd.set_option("display.max_rows", None, "display.max_columns", None)                #To display the entire dataframe
df1=pd.DataFrame({'Job Title':job_titles, 'Job Location':loc, 'Company Name':company_names, 'Experience Required':experience_list})
print("The data for “Data Analyst” Job position in “Bangalore” location")
df1
```

```python
#Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes: 1. Brand 2. Product Description 3. Price

#opening up flipkart.com website on automated chrome window
driver.get('https://www.flipkart.com/')

#finding element in search bar
search_field=driver.find_element_by_class_name("_3704LK") #search bar
search_field.send_keys("sunglasses")

#clicking on search icon
search_icon=driver.find_element_by_xpath("/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/button") #location search
search_icon.click()

brand=[]                    #list to store brand
product=[]                  #list to store product description
price=[]                    #list to store price
discount=[]                 #list to store discount

for j in range (0,3):    
    time.sleep(5)
    #Lets extract all the tags having the brand
    brand_tags=driver.find_elements_by_xpath("//div[@class='_2WkVRV']") #locating web element of brand
    for i in brand_tags:                 
        titles=i.text
        brand.append(titles)
     
    #Lets extract all the tags having the product description
    product_tags=driver.find_elements_by_xpath("//a[@class='IRpwTa']") #locating web element of product description
    
    #Now the text of the product description is inside the tag extracted above
    for i in product_tags:
        products=i.text
        product.append(products)
            
    #Lets extract all the tags having the price
    price_tags=driver.find_elements_by_xpath("//div[@class='_30jeq3']") #locating web element of price
    
    #Now the text of the price is inside the tag extracted above
    for i in price_tags:
        prices=i.text
        price.append(prices) 
   
    #Lets extract all the tags having the discount
    discount_tags=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span") #locating web element of discount
    
    #Now the discount is inside the tag extracted above
    for i in discount_tags:
        discounts=i.text
        discount.append(discounts)
    
    #Now clicking on the next button
    if j==0:                                     #xpath for next butto fo 1st page and upcoming pages are different
        time.sleep(5)
        next_button=driver.find_element_by_xpath("/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]/span")
        next_button.click()
        time.sleep(5)
    else:
        time.sleep(5)
        next_button=driver.find_element_by_xpath("/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[12]/span")
        next_button.click()
        time.sleep(5)

pd.set_option("display.max_rows", None, "display.max_columns", None)                #To display the entire dataframe
df4=pd.DataFrame({'Brand':brand[0:100], 'Product description':product[0:100], 'Price': price[0:100], 'Discount Provided': discount[0:100]})
print("The data for sunglasses-")
df4
```

```python
#Q7. Go to the link - https://www.myntra.com/shoes.Set second Price filter and Color filter to “Black”

#opening up myntra.com website on automated chrome window
driver.get('https://www.myntra.com/shoes')

#applying price filter
driver.get('https://www.myntra.com/shoes?rf=Price%3A6637.0_13125.0_6637.0%20TO%2013125.0')

#applying color filter
driver.get('https://www.myntra.com/shoes?f=Color%3ABlack_36454f&rf=Price%3A6637.0_13125.0_6637.0%20TO%2013125.0')

brand=[]                    #list to store brand
product=[]                  #list to store product description
price=[]                    #list to store price

for j in range (0,2):          
    time.sleep(5)
    #Lets extract all the tags having the brand
    brand_tags=driver.find_elements_by_xpath("//h3[@class='product-brand']") #locating web element of brand
    
    #Now the text of the brand is inside the tag extracted above
    for i in brand_tags:                 
        titles=i.text
        brand.append(titles)
    
    #Lets extract all the tags having the product description
    product_tags=driver.find_elements_by_xpath("//h4[@class='product-product']") #locating web element of product description
    
    #Now the text of the product description is inside the tag extracted above
    for i in product_tags:
        products=i.text
        product.append(products)
            
    #Lets extract all the tags having the price
    price_tags=driver.find_elements_by_xpath("//div[@class='product-price']/span[1]") #locating web element of price
    
    #Now the text of the price is inside the tag extracted above
    for i in price_tags:
        prices=i.text.split()[1]
        pr=prices.replace('Rs.','')
        price.append(''.join(pr))
    
    #Now clicking on the next button
    time.sleep(5)
    next_button=driver.find_element_by_xpath("/html/body/div[2]/div/div[1]/main/div[3]/div[2]/div/div[2]/section/div[2]/ul/li[12]/a")
    next_button.click()
    time.sleep(5)
        
s_no=[]
for k in range(1,101):
    s_no.append(k)

pd.set_option("display.max_rows", None, "display.max_columns", None)                #To display the entire dataframe
df7=pd.DataFrame({'Brand':brand[0:100], 'Product description':product[0:100], 'Price': price[0:100]}, index=s_no)
print("The data for shoes-")
df7
```

```python
#Q10. Write a python program to scrape the salary data for Data Scientist designation. You have to scrape Company name, Number of salaries, Average salary, Minsalary, Max Salary.

#opening up ambitionbox.com website on automated chrome window
driver.get('https://www.ambitionbox.com/')

#clicking on salary option                  
salary_option=driver.find_element_by_xpath("/html/body/div[1]/nav/nav/a[4]") #location search
salary_option.click()

time.sleep(5)
#finding element in search bar
search_field=driver.find_element_by_xpath("//div[@class='searchbox-wrapper']/span/input") #search bar
search_field.send_keys("Data Scientist")

time.sleep(5)
#clicking on Data Scientist option 
ds=driver.find_element_by_xpath("/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p") #location search
ds.click()

company=[]             #list to store company name
tot_sal=[]             #list to store total number ofsalary
avg_sal=[]             #list to store average salary
min_sal=[]             #list to store minimum salary
max_sal=[]             #list to store maximum salary
exp=[]                 #list to store experience required

time.sleep(5)
#Lets extract all the tags having the company names
company_tags=driver.find_elements_by_xpath("//div[@class='name']/a") #locating web element of company name

#Now the text of the company name is inside the tag extracted above
for i in company_tags[0:10]:
    title=i.text
    company.append(title)

#Lets extract all the tags having the total number of salary
tot=driver.find_elements_by_xpath("//div[@class='name']/span") #locating web element of total number of salary

#Now the total no of salary  is inside the tag extracted above
for i in tot[0:10]:
    t=i.text.split()[2]
    tot_sal.append(''.join(t))

#Lets extract all the tags having the average salary
avg=driver.find_elements_by_xpath("//p[@class='averageCtc']") #locating web element of average salary

#Now the average salary is inside the tag extracted above
for i in avg[0:10]:
    r=i.text
    avg_sal.append(r)
    
#Lets extract all the tags having the minimum salary
min=driver.find_elements_by_xpath("//div[@class='salary-values']/div[1]") #locating web element of minimum salary

#Now the minimum salary is inside the tag extracted above
for i in min[0:10]:
    m=i.text
    min_sal.append(m)

#Lets extract all the tags having the maximum salary
max=driver.find_elements_by_xpath("//div[@class='salary-values']/div[2]") #locating web element of maximum salary

#Now the maximum salary is inside the tag extracted above
for i in max[0:10]:
    mx=i.text
    max_sal.append(mx)

#Lets extract all the tags having the experience
exp_tags=driver.find_elements_by_xpath("//div[@class='salaries sbold-list-header']") #locating web element of experience

#Now the experience is inside the tag extracted above
for i in exp_tags[0:10]:
    l=len(i.text.split())
    ex=i.text.split()[3:l-1]
    exp.append(' '.join(ex))

s_no=[]
for k in range(1,11):
    s_no.append(k)
    
pd.set_option("display.max_rows", None, "display.max_columns", None)                #To display the entire dataframe
df10=pd.DataFrame({'Company':company, 'Total Salary Record':tot_sal, 'Average Salary':avg_sal, 'Minimum Salary':min_sal, 'Maximum Salary':min_sal, 'Experience':exp}, index=s_no)
print("The salary for Data Scientist Job”)
df10
```
